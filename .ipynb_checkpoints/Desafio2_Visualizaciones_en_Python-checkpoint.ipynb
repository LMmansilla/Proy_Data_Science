## Dataset N°1 Properati

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt
```

```python
# Se carga archivo CSV
data_properati = pd.read_csv('../ar_properties_1.csv', encoding= "UTF-8")
```

```python
# Se visualizan los registos unicos por regiones. 
prov_cant = data_properati.l2.value_counts()
prov_cant 
```

```python
# Se visualizan diversas provincias que no pertenecen a Argentina, y por ello se procede a filtrarlas. 

data_properati_proc = data_properati[(data_properati['l2'] != 'California') & (data_properati['l2'] != 'Pennsylvania') & (data_properati['l2'] != 'Rio de Janeiro') & (data_properati['l2'] != 'Maryland') & (data_properati['l2'] != 'Santa Catarina') & (data_properati['l2'] != 'Rio Grande do Norte ') & (data_properati['l2'] != 'Miami') & (data_properati['l2'] != 'Rocha') & (data_properati['l2'] != 'São Paulo') & (data_properati['l2'] != 'Michigan') & (data_properati['l2'] != 'Florida') & (data_properati['l2'] != 'Canelones') & (data_properati['l2'] != 'Colonia') & (data_properati['l2'] != 'Maldonado')  & (data_properati['l2'] != 'Rio Grande do Norte')&(data_properati['l2'] != 'Montevideo')]
```

```python
# Se genera un DF con la cantidad de registros por provincia

prov_cant = data_properati_proc.l2.value_counts()
prov_cant_df = pd.DataFrame(prov_cant).reset_index()
prov_cant_df  = prov_cant_df.rename(columns={'index': 'Region', 'l2': 'Cantidad'})
prov_cant_df = prov_cant_df[['Region', 'Cantidad']]
prov_cant_df.sort_values(by = 'Cantidad', ascending=False)
```

```python
# Se genera un grafico para visualizar las cantidades por provincia/región


orden = prov_cant.index
sns.set_style("whitegrid", {'grid.linestyle': '--'})
sns.countplot(data=data_properati_proc, y='l2', order=orden)
plt.xlabel('Cantidad')
plt.ylabel('Provincia o Region')
plt.show()
```

```python
# Se visualiza el tipo de contrato y su cantidad
data_properati_proc.price_period.value_counts()
```

#### Limpieza de datos

```python
# Se filtra los NA de la variable Superficies
data_properati_proc = data_properati_proc.dropna(subset=['surface_total'])

# Se realiza un diagrama de Caja y Bigotes para ver la distribución de la variable Superficies
plt.boxplot(data_properati_proc['surface_total'])
plt.title('Diagrama boxplot')
plt.ylabel('Valores')
plt.show()
```

```python
data_properati_proc.shape
```

```python
# Se realiza una limpieza de outliers de la variable Superficies
Q1 = data_properati_proc['surface_total'].quantile(0.25)
Q3 = data_properati_proc['surface_total'].quantile(0.75)
IQR = Q3 - Q1


limite_inferior1 = Q1 - 1.5 * IQR
limite_superior1 = Q3 + 1.5 * IQR
```

```python
data_properati_2 = data_properati_proc[(data_properati_proc['surface_total'] >= limite_inferior1) & (data_properati_proc['surface_total'] <= limite_superior1) & (data_properati_proc['surface_total'] >= 0)]
data_properati_2.shape
```

```python
# Crear un diagrama de caja y bigotes de la columna 'Valores'

plt.boxplot(data_properati_2['surface_total'])
plt.title('Diagrama boxplot')
plt.ylabel('Valores')
plt.show()
```

### Histograma de superficies

```python
plt.hist(data_properati_2['surface_total'], bins=50, edgecolor='black')
plt.title('Histograma')
plt.xlabel('Superficie')
plt.ylabel('Frecuencia')
plt.show()
```

```python
data_properati_alq = data_properati_2[data_properati_2['operation_type'] == 'Alquiler']

# Se realiza una limpieza de outliers de la variable precio en alquileres

Q1 = data_properati_alq['price'].quantile(0.25)
Q3 = data_properati_alq['price'].quantile(0.75)
IQR = Q3 - Q1


limite_inferior_alq = Q1 - 1.5 * IQR
limite_superior_alq = Q3 + 1.5 * IQR
```

```python
data_properati_alq = data_properati_alq[(data_properati_alq['price'] >= limite_inferior_alq) & (data_properati_alq['price'] <= limite_superior_alq) & (data_properati_alq['price'] >= 0)]
data_properati_alq.shape
```

```python
sns.scatterplot(x='surface_total', y='price', data=data_properati_alq)
plt.xlabel('Superficie')
plt.ylabel('Precio')
plt.title('Scatter Plot')
plt.gca().get_yaxis().get_major_formatter().set_scientific(False)
plt.show()
```
